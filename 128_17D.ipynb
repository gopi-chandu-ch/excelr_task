```python
data={"weight":[66,75,84,96,48,71],"height":[156,165,186,165,174,167]}
bmi=pd.DataFrame(data) #bmi body mass index
bmi
```

```python
# create bmi column "formula = weight(kgs)/height(m-square)"
bmi['BMI']=bmi['weight']/(bmi['height']/100)**2
bmi
```

```python
# insert new record to bmi
bmi.loc[3]=[78,np.NaN,np.NaN]
bmi
```

```python
# here if we give index which is present it will update otherswise it will add
bmi.loc[6]=[np.NaN,65,np.NaN]
bmi
```

```python
bmi.info()
```

```python
bmi.isnull().sum() # if we give only .isnull() it'll show boolean values for the whole table
```

```python
# replace missing value with mean value of the column
bmi["weight"].fillna(65) # .fillna it'll replace null values in weight
bmi
```

```python
bmi["height"]=bmi["height"].fillna(165) # .fillna it'll replace null values in height
bmi
```

```python
bmi['BMI_new']=bmi['weight']/(bmi['height']/100)**2
bmi
```

```python
bmi = bmi.drop("BMI", axis=1) # bmi.drop("BMI", axis=1,inplace=True) if use inplace=True it'll delete permently
bmi
```
