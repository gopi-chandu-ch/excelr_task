```python
import pandas as pd
from sklearn.impute import SimpleImputer
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv('Day_18_Tours_and_Travels.csv')

# Step 2: Handle missing values

# Numerical columns: Impute with median
numerical_columns = ['Package_Price', 'Customer_Age', 'Rating']
numerical_imputer = SimpleImputer(strategy='median')
df[numerical_columns] = numerical_imputer.fit_transform(df[numerical_columns])

# Categorical columns: Impute with mode (most frequent value)
categorical_columns = ['Destination', 'Review_Text', 'Travel_Date']
categorical_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_columns] = categorical_imputer.fit_transform(df[categorical_columns])

# Step 3: Verify no missing values
print(df.isna().sum())

# Step 4: Handle duplicates
df = df.drop_duplicates()

# Step 5: Detect outliers with boxplots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Package_Price')
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='Rating')
plt.tight_layout()
plt.show()

# Step 6: Save cleaned data
df.to_csv('Day_18_Tours_and_Travels.csv', index=False)

print("Data cleaned and saved.")
```
