```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "Day_12_banking_data.csv"
banking_data = pd.read_csv(file_path)

# 1. Plot the total sum of Transaction_Amount per Account_Type using a bar plot
total_transaction_by_account_type = banking_data.groupby("Account_Type")["Transaction_Amount"].sum()
total_transaction_by_account_type.plot(kind='bar', color='skyblue', figsize=(8, 6))

plt.title("Total Transaction Amount per Account Type")
plt.xlabel("Account Type")
plt.ylabel("Total Transaction Amount")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# 2. Create a pie chart to show the percentage of transactions per Branch
transaction_per_branch = banking_data.groupby("Branch")["Transaction_Amount"].sum()
transaction_per_branch.plot(kind='pie', autopct='%1.1f%%', figsize=(7, 7), colors=['lightcoral', 'lightgreen', 'lightblue', 'orange'])

plt.title("Percentage of Transactions per Branch")
plt.ylabel('')  # Remove the default label on the y-axis
plt.show()
```
