```python
import pandas as pd

file_path = "Day_8_sales_data.csv"
sales_data = pd.read_csv(file_path)

# 1. Extract all rows where sales are greater than 1000
sales_greater_than_1000 = sales_data[sales_data["Sales"] > 1000]
print("Rows where sales are greater than 1000:")
print(sales_greater_than_1000, "\n")

# 2. Find all sales records for a specific region (e.g., "East")
region_sales = sales_data[sales_data["Region"] == "East"]
print('Sales records for the region "East":')
print(region_sales)
```

```python
import pandas as pd

file_path = "Day_8_sales_data.csv" 
sales_data = pd.read_csv(file_path)

# 1. Add a new column, Profit_Per_Unit, calculated as Profit / Quantity
sales_data["Profit_Per_Unit"] = sales_data["Profit"] / sales_data["Quantity"]
print("DataFrame after adding the Profit_Per_Unit column:")
print(sales_data, "\n")

# 2. Create another column, High_Sales, labeling rows as Yes if Sales > 1000, else No
sales_data["High_Sales"] = sales_data["Sales"].apply(lambda x: "Yes" if x > 1000 else "No")
print("DataFrame after adding the High_Sales column:")
print(sales_data)
```
